```python
# import the pandas library and set an object to it called pd
import pandas as pd
```

```python
#open the assignment
data = pd.read_csv('Realestate.csv')
```

```python
#open the top key rows
data.head()
```

```python
#set a variable to the house age of a row
houseage = data.houseAge
```

```python
#Step 2 - Normalize house age col using standardization
norm = (houseage - houseage.mean())/houseage.std()   ## Called standardization in statistics
norm
```

```python
#open the data (real estate file) data frame.
data
```

```python
#step 3 -create new column
data['standardized_houseage'] = norm
```

```python
data # display result
```

```python
#step 4 drop column
data.drop('numberOfConvenienceStores',axis=1,inplace=True)
# or data.drop(columns=[numberOfConvenienceStores])
#or data.drop(columns=[numberOfConvenienceStores],inplace=True)
```

```python
data
```

```python
# Step 5 column rename transaction to transactionDate
#Before
data.columns
#after inplace=True locks it down in the original Data Frame
data.rename(columns={"transaction": "transactionDate"}, inplace = True)

#after rename
#check the results
data.columns

```

```python
data  #review file
```

```python
# 6 Use loc to display all rows from 0 through 10

data.loc[0:10, :]
```

```python
#7  iloc - reference by index number --last row is exclusive
data.iloc[0:10, :]
```

```python
#8  drop data and keep the first row where the data duplicates
data.drop_duplicates(keep='first',inplace=True)
```

```python
data.count()  # reduced from 418
```

```python
data.columns
```

```python
#Step 9  Fill in missing as means
#data.fillna(data.mean())
data.fillna(value={'No': data['No'].mean(),'transactionDate': data['transactionDate'].mean()
, 'houseAge': data['houseAge'].mean() , 'distanceToNearestMRTstation': data['distanceToNearestMRTstation'].mean()
,'latitude': data['latitude'].mean() 
,'longitude': data['longitude'].mean() 
,'housePriceOfUnitArea': data['housePriceOfUnitArea'].mean()                  
,'standardized_houseage': data['standardized_houseage'].mean()                
}, inplace=True)                 
              




```
