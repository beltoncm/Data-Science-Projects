```python

#import necessary libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#pull in the main file
data = pd.read_csv('countries.csv')
data.head()
```

```python

#Create a slice for all records where the year is 1952
active_statuses = [1952]
mask = data['year'].isin(active_statuses)
active_data = data[mask]

print(active_data)

active_data
```

```python
#slice the data set further by pulling in a frame for Africa 1952 records
africa_1952 = active_data[active_data.continent == 'Africa']

#slice the data set further by pulling in a frame for Asia 1952 records
asia_1952 = active_data[active_data.continent == 'Asia']
```

```python
#check 1952 data set
active_data.head()
```

```python
#check the first five rows to see if it is records with year of 1952 and value of Africa

africa_1952.head()
```

```python
#check the first five rows to see if it is records with year of 1952 and value of Asia

asia_1952.head()
```

```python
#check row count for africa 1952

africa_1952.shape

#check row count for asia 1952


asia_1952.shape
```

```python

## CREATE A ROW NUMBER TO SERVE AS an observation number for the x axis for both the Africa and Asia data frame

africa_1952 = africa_1952.assign(row_number=range(len(africa_1952)))

asia_1952 = asia_1952.assign(row_number=range(len(asia_1952)))


#create subplt 2 rows 2 columns 1st chart
plt.subplot(2,2,1)
#plot observation number as x and life expectancy as y for Africa
plt.plot(africa_1952.row_number,africa_1952.lifeExpectancy)
#set numbering limits on the x axis
plt.xlim(1,70)
#plt.set_xticks(range(1,70, 5))
plt.xlabel('observation')
plt.title('Life Expectancy for Africa (1952)')
#plt.ylabel('Life Expectancy')

#create subplt 2 rows 2 columns 2nd chart
plt.subplot(2,2,2)
plt.plot(asia_1952.row_number, asia_1952.lifeExpectancy)
#set numbering limits on the x axis
plt.xlim(1,70)
#plt.legend(['Asia', 'Africa'])
plt.xlabel('observation')
plt.title('Life Expectancy for Africa (1952)')
#plt.ylabel('Life Expectancy')
plt.show()
```

```python
## COMMENT UP SHOULD BE GOOD TO SUBMIT
#create subplt 2 rows 2 columns 1st chart
plt.subplot(2, 1, 1) # subplot(211)
#chart title
plt.title('Life Expectancy')
#histogram for asia life expectancy in 1952, number of bins 20 and max range 100
plt.hist(asia_1952.lifeExpectancy, 20, range=(0, 100), edgecolor='black')
#add y label
plt.ylabel('Asia')

plt.subplot(2, 1, 2)
#histogram for asia life expectancy in 1952, number of bins 20 and max range 100
plt.hist(africa_1952.lifeExpectancy, 20, range=(0, 100), edgecolor='black')
plt.ylabel('Africa')
plt.show()
```
